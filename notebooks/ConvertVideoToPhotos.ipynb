```python
import cv2
import os

# Opens the inbuilt camera of laptop to capture video.
video_path = '/data/aiscience/IMG_0708.MOV'
video_name = os.path.splitext(os.path.split(video_path)[-1])[-2]
print(video_name)
cap = cv2.VideoCapture(video_path)

i = 0

output_folder = f"/data/pixtrack/pixel-perfect-sfm/datasets/MotorCore_06June2022/query/{video_name}"
os.mkdir(output_folder)
while(cap.isOpened()):
    ret, frame = cap.read()
    # This condition prevents from infinite looping
    # incase video ends.
    if ret == False:
        break
    image_path = os.path.join(output_folder, f"{video_name}_{str(i).zfill(4)}.png")
    # Save Frame by Frame into disk using imwrite method
    cv2.imwrite(image_path, frame)
    print(f"writing to: {image_path}")# frame)
    i += 1
 
cap.release()
cv2.destroyAllWindows()
```
